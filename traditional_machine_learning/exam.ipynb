```python
corr_dic={}
for column in corr_matrix.columns:
    corr_temp=[]
    for feature, value in corr_matrix[column].iteritems():
        if abs(value)>0.85 and feature!=column:
            corr_temp.append((feature,value))
    if len(corr_temp)>0:
        corr_dic.update({column:corr_temp})
print(corr_dic)
```

```python
# find outliers
Q1 = wine_dataset_corr.quantile(0.25)
Q3 = wine_dataset_corr.quantile(0.75)
IQR = Q3 - Q1
```

```python
wine_dataset_out = wine_dataset_corr[~((wine_dataset_corr < (Q1 - 1.5 * IQR)) |(wine_dataset_corr > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python

plt.subplots(1, 3, figsize=(18, 8))
plt.subplot(1, 3, 1)
sns.boxplot(data=wine_dataset_out, x='target', y='alcohol')
plt.plot()
plt.subplot(1, 3, 2)
sns.boxplot(data=wine_dataset_out, x='target', y='malic_acid')
plt.plot()
plt.subplot(1, 3, 3)
sns.boxplot(data=wine_dataset_out, x='target', y='ash')
plt.plot()
plt.show()
```

```python
X=wine_dataset_out.drop(['target'],axis=1)
y=wine_dataset_out['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report,f1_score
f_score_arry=[]
def evaluate(grid_search,y_test, y_pred):
    # Confusion matrix
    print("Confusion Matrix")
    matrix = confusion_matrix(y_test, y_pred)
    print(matrix)
    # Classification Report
    print("\nClassification Report")
    report = classification_report(y_test, y_pred)
    print(report)
    # Accuracy of the model
    accuracy = accuracy_score(y_test, y_pred)
    print('Accuracy {:.2f}%'.format(accuracy*100))
    print()
    print("Best parameters set grid search:")
    print(grid_search.best_params_)
    f_score=f1_score(y_test, y_pred,average='weighted')
    return f_score
```

```python
from sklearn.model_selection import GridSearchCV
def predictor(pipe,search_hyper_params,X_train, y_train):
    grid_search = GridSearchCV(pipe, search_hyper_params, cv=10, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    y_pred = grid_search.predict(X_test)
    return grid_search,y_pred
```

```python
from sklearn import preprocessing
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
import time
```

```python
# decision_function_shape='ovr'------one-vs-rest
svm_classifier = SVC(kernel='rbf')
pipe_steps = [('scaler', preprocessing.StandardScaler()), ('svc', svm_classifier)]
pipe = Pipeline(steps=pipe_steps)
```

```python
search_hyper_params = {'svc__C': [0.01, 0.1, 1 , 8, 10 , 100, 1000], 'svc__gamma': [10,1, 0.1, 0.01, 0.001, 0.0001]}
```

```python
t0_ovr = time.time()
model,y_pred=predictor(pipe,search_hyper_params,X_train, y_train)
t1_ovr = time.time()
```

```python
print('*'*50)
print(' '*10 + 'svc (kernel= rbf)--one-vs-rest')
print('*'*50)
f_score= evaluate(model,y_test, y_pred)
f_score_arry.append(['svc (kernel= rbf)--one-vs-rest',f_score])
```

```python
# decision_function_shape='ovo'------one-vs-one
svm_classifier = SVC(kernel='rbf',decision_function_shape='ovo')
pipe_steps = [('scaler', preprocessing.StandardScaler()), ('svc', svm_classifier)]
pipe = Pipeline(steps=pipe_steps)
```

```python
t0_ovo = time.time()
model,y_pred=predictor(pipe,search_hyper_params,X_train, y_train)
t1_ovo = time.time()
```

```python
print('*'*50)
print(' '*10 + 'svc (kernel= rbf)--one-vs-one')
print('*'*50)
f_score=evaluate(model,y_test, y_pred)
f_score_arry.append(['svc (kernel= rbf)--one-vs-one',f_score])
```

```python
print('model training in svc (kernel= rbf)--one-vs-rest: ', t1_ovr-t0_ovr)
print('model training in svc (kernel= rbf)--one-vs-one: ', t1_ovo-t0_ovo)
```

```python
from sklearn.svm import LinearSVC

linearSVC_classifier=LinearSVC(random_state=42)
pipe_steps = [('scaler', preprocessing.StandardScaler()), ('linearSVC', linearSVC_classifier)]
pipe = Pipeline(steps=pipe_steps)
search_hyper_params = {'linearSVC__C': [0.001,0.01,0.1,1,10,100,1000]}
t0 = time.time()
model_1,y_pred_1=predictor(pipe,search_hyper_params,X_train, y_train)
t1 = time.time()
print('model training in LinearSVC: ', t1-t0)

svm_classifier = SVC(kernel='linear',decision_function_shape='ovr')
pipe_steps = [('scaler', preprocessing.StandardScaler()), ('svc', svm_classifier)]
pipe = Pipeline(steps=pipe_steps)
search_hyper_params = {'svc__C': [0.001,0.01,0.1,1,10,100,1000]}
t0 = time.time()
model_2,y_pred_2=predictor(pipe,search_hyper_params,X_train, y_train)
t1 = time.time()
print('model training in svc (kernel=linear,ovr): ', t1-t0)

svm_classifier = SVC(kernel='linear',decision_function_shape='ovo')
pipe_steps = [('scaler', preprocessing.StandardScaler()), ('svc', svm_classifier)]
pipe = Pipeline(steps=pipe_steps)
search_hyper_params = {'svc__C': [0.001,0.01,0.1,1,10,100,1000]}
t0 = time.time()
model_2,y_pred_2=predictor(pipe,search_hyper_params,X_train, y_train)
t1 = time.time()
print('model training in svc (kernel=linear,ovo): ', t1-t0)
```

```python
print('*'*50)
print(' '*10 + 'LinearSVC')
print('*'*50)
f_score=evaluate(model_1,y_test, y_pred_1)
f_score_arry.append(['LinearSVC',f_score])
```
